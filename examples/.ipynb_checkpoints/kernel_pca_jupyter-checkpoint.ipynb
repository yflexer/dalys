# Quick tutorial to Dalys

```python
%matplotlib inline
import matplotlib.pyplot as plt
from dalys.kernel_pca import KPCATool
from sklearn.datasets import make_circles, make_moons, load_digits, load_iris
```

## For dataset example we take scikit-learn [circles dataset](http://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_circles.html) 

```python
x, y = make_circles(n_samples=1000, random_state=123, noise=0.1, factor=0.2)

plt.scatter(x[y == 0, 0], x[y == 0, 1], color='red')
plt.scatter(x[y == 1, 0], x[y == 1, 1], color='blue')
plt.show()
```

## Let's demonstrate the work of the Kernel PCA algirithm on the current dataset

```python
style = [('red', '.'), ('blue', '.')]
label_names = ['red points', 'blue points']
kpca = KPCATool(x, y, style=style, labels_unique_name=label_names, n_components=3, kernel='rbf', gamma=2, scaled=True)
kpca.projections_plot(grid=22)
```

## If you want to visualize the rbf kernel, you can use ***projections_plot_3d*** method:

> Method ***projections_plot_3d*** has argument "*componets*" who tell you what components must be visualised.

Example:

```python
kpca.projections_plot_3d(components=(0, 1, 2))
```

There is a number of parameters that can be set for the KPCATool class;
the major ones are as follows:

* *style* - set marker styles and colors for data representation, 
example: *style* = [('red', '.'), ('blue', '.')] or *style* = [('red', 'd'), ('blue', 'd')]. If you need random colors and markers - leave this argument with default parameter (default = None).

* *labels_unique_name* - set list with class names, such as *labels_unique_name* = ['red dotes', 'blue dotes']. If *labels_unique_name* is *None*, this means that labels will be numbered in ascending order (default = None).

* *scaled* - if your data is already transformed(normalized, standardize or whitening and so all) - you must set "True" for this flag to prevent data preprocessing,
if you need to standardize your data - you must use tools with default argument (default = False).

* *scale_axis* - set axis for data standardize (default = 0)
