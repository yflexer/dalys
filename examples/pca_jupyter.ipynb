# Quick tutorial to Dalys

## *Introduce to PCA*

```python
%matplotlib inline
import matplotlib.pyplot as plt
from dalys.pca import PCATool
from sklearn.datasets import load_digits
```

## For dataset example we take scikit-learn [digits dataset](http://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_digits.html)

```python
digits = load_digits()
x = digits.data
y = digits.target

plt.figure(1, figsize=(5, 5))
plt.imshow(digits.images[0], cmap=plt.cm.gray_r, interpolation='nearest')
plt.show()
```

## Let's demonstrate the work of the PCA algorithm on the [digits dataset](http://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_digits.html)

```python
names = ['zero', 'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine']
styles = [('xkcd:neon blue', ','), ('xkcd:yellowish tan', '+'),
          ('xkcd:cornflower blue', '+'),('xkcd:fire engine red', 'D'),
          ('xkcd:brownish red', '8'), ('darkgoldenrod', 'v'),
          ('xkcd:orangered', 'v'), ('xkcd:faded green', 'p'),
          ('xkcd:dark royal blue', 's'), ('xkcd:lemon yellow', '3')]
pca = PCATool(x, y, style=styles, labels_unique_name=names, n_components=3, preprocessing='std')
pca.projections_plot(grid=22)
```

## Large plots with legend

```python
pca.projections_plot()
```

## 3D plot

```python
pca.projections_plot_3d()
```
